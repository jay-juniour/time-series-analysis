# Time Series Analysis(TSA) 

Time Series Analysis (TSA) is a method of investigating the attributes of the dependent variable over time in relation to the response variable,also is a specific way of analyzing a sequence of data points collected over an interval of time.   We draw on the time variable as a benchmark when estimating the target variable in forecasting . The time intervals represented by TSA are in the following order of magnitude: years, months, weeks, days, hours, minutes, and seconds. It is an observation derived from the succession of discrete time intervals.

This makes it a stochastic process beacuse a stochastic process is a collection of random variables indexed by time or another parameter.

Time series Analysis (TSA) is vital to prediction, specific to time-based problem statements.

## Basic concepts in Time Series
A lag is a period of time between one event and another.
A lag in TSA is the observation at time stamp k, $y {_k-i}$ (i periods behind in time) is referred to as the $i$-th lag of $y _k$.

White noise error in time series is a kind of random variability that has an impact on the observations, they cannot be attributed to any systematic pattern. They have a mean of zero and a constant variance over time.

## Components of Time Series Analysis
Trend: Trend shows the pattern of the time series over a period of time, whether increasing, decreasing or constant.

Seasonality: Indicates whether a time series exhibits seasonal effects, which are characterised by consistent shifts or fixed intervals.

Cyclical: When a time series exhibits pattern and movement uncertainty in the absence of a fixed interval

Irregularity: Unexpected events and spikes in a short time span.

## Time Series Data Types¬∂
There are two major data types - stationary and non-stationary.

## Stationary
A time series is said to be stationary if its statistical properties such as variance, standard deviation, mean, covariance and variance do not vary with time, or are not a function of time. In simpler terms, a series without seasonal or trend components. In order to avoid spurious correlations and regressions, a time series should be stationary. The study shall employ the Augmented Dickey-Fuller Test to test for stationarity.

## Non-Stationary
When either the mean, variance or covariance changes with respect to time. The time series is said to be non-stationary.

## Tests for Stationarity
Stationarity can be checked using the following statistical tests:
 + Augmented Dickey-Fuller (ADF) Test
 + kwiatkowski-Phillips-Schmidt-Shin (KPSS) Test
The study shall employ the ADF test.

#### Augmented Dickey-Fuller (ADF) Test
This is a unit root statistical test. A unit root is a characteristic of certain stochastic processes in probability theory and statistics, which can complicate statistical inference involving time series models. To put it simply, while the unit root does not always have a trend component, it is non-stationary.

The following assumptions are made when conducting the ADF:

 + Null Hypothesis ( ùêª0 ): Series is non-stationary, or series has a unit root.
 + Alternate Hypothesis( ùêªùëé): Series is stationary, or series has no unit root.
 If the null hypothesis fails to be rejected, then there is evidence that the time series is non-stationary.

Conditions to reject H0:
  + If the test statistic < critical value or p-value < 0.05. The structure of the time series is not time-dependent.

## Cross-Correlation

The study explores cross-correlation instead of simple correlation since the simple correlation coefficient will not wholly utilize the data, as a result some important information may remain hidden since it compares pairs of two time series at the same time stamp.

Cross-correlation is an extremely useful instrument for identifying patterns and dependencies between time series variables, as it calculates the degree of similarity between two time series in relation to the time lag that is applied to one of them. A time series that has a 'k'th lag, is basically like stating that there is a version of the time series which is 't-k' periods behind in time, lag (k=1) is a version of the initial time series that is behind time by 1 period. It is determined by sliding one time series over the other and calculating the correlation coefficient at each interval. This shift is done in such a manner that say for instance we have two series x and y. Where x consists of [1, 3, 5, 7] and y [4, 6, 8, 10] and relative to y we shift x to the right by one data point then match the two series, 4 in y will not be matched. In our case, we are investigating the possibilty that the current values of weather affect future values of coffee yield.

Cross-correlation relies on the units of x and y. There is difficulty when comparing the cross-correlations whose units are different. For example temperature in degrees celcius and yield in number of buckets, this would cause the sum of products decrease as the lags increase. This problem is solved by normalizing the data to ensure that all the values of the variables x and y used in calculating cross-correlation. The normalization is done using the the MinMaxScaler from scikit-learn which can be found [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html) which scales the values to a range between 0 and 1 and the formula that has been used in that computation is as follows:
\begin{equation}
X _{scaled} = \frac{X-X _{min}}{X _{max} - X _{min}}
\end{equation}

where:
* $X$ is the original data point.
* $X _{min}$ is the minumum value of the variable
* $X _{max}$ is the maximum value of the variable

The cross correlation function is calculated using the following formula:

\begin{equation}
R _{xy} = \frac{\sum _{t=1} ^{N-k} (x[t] - \bar{x}) (y[t+k] - \bar{y})} {\sqrt{\sum _{t=1} ^ N (x[t] - \bar{x})^2 \sum _{t=1} ^N (y[t]-\bar{y})^2}}
\end{equation}

where:
* $R _{xy}[k]$ is the cross-correlation at lag k
* $x[t]$ and $y[t]$ are the time series values as time t for the two variables being correlated.
* $N$ is the otal number of observations in the time series.
* $\bar{x}$ and $\bar{y}$ are the means of the respective time series.

In this case we use positive lags because in so doing, we shift the values of temperature forward in time relative to those of yield. When we use k lags, it means that we compare the current values of temperature with the values of yield k time units (months) in the future.

### Correlating the Moving Average of Precipitation with yield
The study, in an attempt to find what would result in a strong positive cross-correlation between precipitation and yield, will obtain the moving average of precipitation and use it to get the cross-correlation with yield.

The formula adopted for the moving average is:
\begin{equation}
sma = \frac{x _1 + x _2 + ... + x _n}{n}
\end{equation}

where:
* sma is the simple moving average
* $x _1, x _2, ..., x _n$ are the values of the time series over n periods
* $n$ is the number of periods (window size)

This is the implementation used by [pd.rolling](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rolling.html) in pandas.

```python
# Cross correlate the moving average of rainfall and yield
window_size=3

# Get the moving averages
rainfall_ma = prec_monthly_sum.rolling(window=window_size, min_periods=1).sum()
rainfall_ma_yield = rainfall_ma.merge(y_df, on="date", how="inner")

# Correlate with yield
prec_ma_ts = TimeSeries(rainfall_ma_yield, "yield")
ma_ccf, ma_ci = prec_ma_ts.cross_correlate(rainfall_ma_yield["precipitation (mm)"], rainfall_ma_yield["yield"], 12)
ma_ccf_plot = prec_ma_ts.plot_ccf(ma_ccf, ma_ci, suppress_labels=True)
```

#### Observation
Even with the moving average (which considers the moving average of rainfall, there is still a very weak correlation with yield.
